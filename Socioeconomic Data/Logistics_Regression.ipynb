```python
import pandas as pd
import numpy as np
from sklearn import metrics
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
data1 = pd.read_csv('finaldata.csv')
```

```python
data1.head(10)
```

```python
X = data1
toKeep = ['DomesticMigration', "Births", 'Unemployed', "actual_arson_storage","actual_assault_unarmed","actual_burg_force_entry","actual_rape_attempted","actual_robbery_unarmed","actual_robbery_with_a_gun","Poverty"]
Y = data1['Most Likely HT Binary_1']

for i in data1.columns:
    if i in toKeep:
        continue
        
    else:
        X = X.drop(i, axis = 1)
```

```python
# logistic regs

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.20)

logReg = LogisticRegression( max_iter = 1000)
logReg.fit(x_train, y_train)

score = logReg.score(x_test, y_test)
print(score)
```

```python
pred = logReg.predict(x_test)
cm = metrics.confusion_matrix(y_test, pred)
print(cm)
```
